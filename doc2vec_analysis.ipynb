```python
from gensim.models.doc2vec import Doc2Vec
```

# Doc2vec를 이용하여 할 수 있는 여러 기능들에 대해 정리 해놓은 파일이다.
https://radimrehurek.com/gensim/models/doc2vec.html

```python
model_name = 'dynamic_analysis\\Doc2vec_model_vector30_window15_dm0'
model = Doc2Vec.load(model_name) #저장된 모델 부르기
```

```python
model.wv.index2word#단어의 빈도수 순서대로 나열됨.
```

```python
model.docvecs.doctags.keys()#모델에 붙인 태그를 보여줌.
```

```python
model.docvecs.vectors_docs#각 태그들의 vector 값들을 보여줌.
```

```python
model.corpus_count#그룹 개수를 보여줌.
```

```python
model.wv.vocab#모델의 단어들을 보여줌.
```

```python
model.docvecs.doctags# tag 을 딕셔너리 형태로 가져옴.
```

```python
model.docvecs.index_to_doctag(5000)#docmodel에 저장되어있는 5000번째 배열의 tag를 가져옴
```

```python
#doc2vec 모델을 사용하여 추론을 통해 벡터로 변환
vector = model.infer_vector(['c','b','c','d','e','f','g'],
                            alpha=0.1,min_alpha=0.01, epochs=1)
```

```python
vector
```

```python
model.docvecs.most_similar([vector]) #벡터와 가장 비슷한 tags 를 반환함.
```
